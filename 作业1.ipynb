#### 数据分析 (analytics)

```python

df["Berkshire_Hathaway_return_value"] = (df.Berkshire_Hathaway_return+1).cumprod().values
```

```python
df["BRK_13_F_public_stocks_holding_return_value"] = (df.BRK_13_F_public_stocks_holding_return+1).cumprod().values
```

```python
df["risk_free_return_value"] = (df.risk_free_return+1).cumprod().values
```

```python
df["market_excess_return_value"] = (df.market_excess_return+1).cumprod().values
```

```python
df["SMB_value"] = (df.SMB+1).cumprod().values
```

```python
df["HML_value"] = (df.HML+1).cumprod().values
```

```python
df['Berkshire_Hathaway_return_value_HWM'] = df.Berkshire_Hathaway_return_value.cummax()
```

```python
df['drawdown'] = (df.Berkshire_Hathaway_return_value_HWM - df.Berkshire_Hathaway_return_value)/df.Berkshire_Hathaway_return_value_HWM
```

```python
fig = plt.figure() 
ax = fig.add_subplot(211)
df[['Berkshire_Hathaway_return_value','Berkshire_Hathaway_return_value_HWM']].plot(ax=ax, figsize=(16,8)) 

ax = fig.add_subplot(212)
df[['drawdown']].plot(ax=ax, color="green", figsize=(16,8))
```

```python
fig.savefig("Berkshire_Hathaway_return_value_HWM.pdf")
```

```python
df['BRK_13_F_public_stocks_holding_return_value_HWM'] = df.BRK_13_F_public_stocks_holding_return_value.cummax()
```

```python
df['drawdown'] = (df.BRK_13_F_public_stocks_holding_return_value_HWM - df.BRK_13_F_public_stocks_holding_return_value)/df.BRK_13_F_public_stocks_holding_return_value_HWM
```

```python
fig = plt.figure() 
ax = fig.add_subplot(211)
df[['BRK_13_F_public_stocks_holding_return_value','BRK_13_F_public_stocks_holding_return_value_HWM']].plot(ax=ax, figsize=(16,8)) # 告诉pandas现在把图画在当前坐标ax

ax = fig.add_subplot(212)
df[['drawdown']].plot(ax=ax, color="green", figsize=(16,8))
```

```python
fig.savefig("BRK_13_F_public_stocks_holding_return_value_HWM.pdf")
```
